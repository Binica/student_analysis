Using the similarity metrics (Euclidean distance and Manhattan
distance), write code to find out the 5 most similar students to the
students with studentIDs GP23 and MS46. The choice of the attributes
to calculate the similarities is up to you.

```python
# import the relevant libraries
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.metrics.pairwise import manhattan_distances
```

```python
# choose the attributes to find similarity
attributes = ['absences', 'G1', 'G2','G3']
# assign all the values of the choosen atrributes to GP23 variable
GP23 = new_dataset[new_dataset['studentID'] == 'GP23'][attributes]
# assign all the values of the choosen attributes to MS46 variable
MS46 = new_dataset[new_dataset['studentID'] == 'MS46'][attributes]
```

## Using Euclidean distance

```python
# define empty dictionaries
GP23_euc = {}
MS46_euc = {}
# use for loop to calculate Euclidean distances with each students and store the results in dictionary
for each in new_dataset['studentID']:
    id_details = new_dataset[new_dataset['studentID'] == each][attributes]
    GP23_euc_result = euclidean_distances(GP23,id_details)
    GP23_euc[each] = GP23_euc_result[0][0]
    MS46_euc_result = euclidean_distances(MS46,id_details)
    MS46_euc[each] = MS46_euc_result[0][0]
```

```python
# sort the distances to find the 5 most similar students with GP23. The students with least distances are more similar.
{k: v for k, v in sorted(GP23_euc.items(), key=lambda item: item[1])[0:6]}
```

```python
# sort the distances to find the 5 most similar students with MS46. The students with least distances are more similar.
{k: v for k, v in sorted(MS46_euc.items(), key=lambda item: item[1])[0:6]}
```

## Using Manhattan distance

```python
# define empty dictionaries
GP23_mht = {}
MS46_mht = {}
# use for loop to calculate Manhattan distances with each students and store the results in dictionary
for each in new_dataset['studentID']:
    id_details = new_dataset[new_dataset['studentID'] == each][attributes]
    GP23_mht_result = manhattan_distances(GP23,id_details)
    GP23_mht[each] = GP23_mht_result[0][0]
    MS46_mht_result = manhattan_distances(MS46,id_details)
    MS46_mht[each] = MS46_mht_result[0][0]
```

```python
# sort the distances to find the 5 most similar students with GP23. The students with least distances are more similar.
{k: v for k, v in sorted(GP23_mht.items(), key=lambda item: item[1])[0:6]}
```

```python
# sort the distances to find the 5 most similar students with MS46. The students with least distances are more similar.
{k: v for k, v in sorted(MS46_mht.items(), key=lambda item: item[1])[0:6]}
```
